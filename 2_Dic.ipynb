```
def fibonaci(n):
  print("llamada",n)
  if n== 1 or n ==0:
    return n
  else:
    return (fibonaci(n-1) + fibonaci (n-2))

print(fibonaci(7))
```

```
def merge_sort(arreglo, izq,der):
  if izq>=der:
    return
  mitad=  (izq + der)//2
  merge_sort (arreglo, izq, mitad)
  merge_sort(arreglo, mitad+1, der)
  print("Llamar Merge")
  merge_sort(arreglo,izq,der,mitad)
```
